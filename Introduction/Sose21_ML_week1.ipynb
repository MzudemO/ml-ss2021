```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt

import random, math, glob, statistics

# Importing Image and ImageOps module from PIL package
from PIL import Image, ImageOps
```

## Read Data

for IRIS flower dataset(https://archive.ics.uci.edu/ml/datasets/Iris)


```python
# Read single data file
file = "iris.txt"
sepal_length, sepal_width, petal_length, = np.loadtxt(file, usecols=(0,1,2), skiprows=1, delimiter=',', unpack=True)
species = labels = np.loadtxt(file, skiprows=1, dtype=str, delimiter=',', usecols=(4))
print(sepal_length, sepal_width, species)
```

## Plot Data

## Visualization with matplotlib

Cheatsheet link: https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Matplotlib_Cheat_Sheet.pdf

```python
# Set up the figure
f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

for g in np.unique(species):
    i = np.where(species == g)
    ax.scatter(sepal_length[i], sepal_width[i], label=g)

ax.set_xlabel("sepal length", fontsize=15)
ax.set_ylabel("sepal width", fontsize=15)
ax.set_title('IRIS flower data', fontsize=20)
ax.legend()

plt.show()
```

## Visualization with seaborn

Cheatsheet link: https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Seaborn_Cheat_Sheet.pdf

```python
import seaborn as sns
 
iris = sns.load_dataset('iris')
 
# style used as a theme of graph
# for example if we want black
# graph with grid then write "darkgrid"
sns.set_style("whitegrid")
 
# sepal_length, petal_length are iris 
# feature data height used to define
# Height of graph whereas hue store the
# class of iris dataset.
sns.FacetGrid(iris, hue ="species",
              height = 6).map(plt.scatter,
                              'sepal_length',
                              'petal_length').add_legend()
```

## Function to load every jpg image in a given path

```python
# Function to load every png image in a given path
def load_data(cls):
    for filename in glob.glob(cls + "/*.jpg"):
        yield Image.open(filename)

x0 = list(load_data('images/mnist_test_sample'))
random.shuffle(x0)

# mean of grayscaled image
def grayscale_mean(img):
    return statistics.mean(img.getdata())

plt.imshow(x0[0])

print(x0[0])

print(grayscale_mean(x0[0]))
```

## Loading images from archive file (from MNIST dataset)

Handwritten digits dataset (http://yann.lecun.com/exdb/mnist/)

```python
import gzip
mnist_images = gzip.open('t10k-images-idx3-ubyte.gz','r')

image_size = 28
num_images = 5

mnist_images.read(16)
buf = mnist_images.read(image_size * image_size * num_images)
mnist_data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
mnist_data = mnist_data.reshape(num_images, image_size, image_size, 1)

image = np.asarray(mnist_data[3]).squeeze()
plt.imshow(image)
plt.show()
```
